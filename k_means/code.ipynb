Let's implement K-Means from scratch using only NumPy — no sklearn. Every line maps directly to the math we just built.

### Fulll implementation

#### Step 1 - Distance calculation

```python
import numpy as np

def compute_distances(X, centroids):
    
    distances = np.zeros((X.shape[0], centroids.shape[0]))
    
    for k, mu in enumerate(centroids): 
        diff = X - mu  # (n, d)
        distances[:, k] = np.sum(diff**2, axis=1)  # squared euclidean
    
    return distances 
```

#### Step 2 - Assignment

```python
def assign_clusters(X, centroids):
    distances = compute_distance(X, centroids)
    return np.argmin(distances, axis=1) 
```

#### Step 3 - Centroid Update

```python
def update_centroids (X, labels, K): 
    # X: (n, d)
    # labels: (n,) → index of closest centroid for each point
    # K: number of centroids 

    d = X.shape[1]
    centroids = np.zeros((K, d))

    for k in range(K): 
        points = X[labels == k] 
        centroids[k] = points.mean(axis=0)

    return centroids
```
